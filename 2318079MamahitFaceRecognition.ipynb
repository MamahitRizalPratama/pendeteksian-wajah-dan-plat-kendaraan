## Deteksi banyak wajah dan kaburkan

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

digier = os.path.join(path, "data/rizalpkkmbgray.jpg")
img = cv2.imread(digier, 0)

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

deteksi_wajah = face_cascade.detectMultiScale(
    img,
    minNeighbors=4,
    scaleFactor=1.05,
    maxSize=(40,40)
)

for (x, y, w, h) in deteksi_wajah:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255), 3)

plt.figure(figsize=(15,10))
plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
many = os.path.join(path, "data/rizalpkkmbgray.jpg")
img = cv2.imread(many, 0)

face_cascades = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# deteksi wajah
deteksi_wajah = face_cascades.detectMultiScale(
    img,
    scaleFactor=1.04,
    minNeighbors=8,
    minSize=(50,50)
)

# gambar kotak
for (x, y, w, h) in deteksi_wajah:
    cv2.rectangle(
        img,
        (x, y),
        (x+w, y+h),
        (0,0,0),
        30
    )

# tampilkan gambar
plt.figure(figsize=(15,10))
plt.axis('off')
plt.imshow(img, cmap='gray')
```
